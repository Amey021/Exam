```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero"
    return a / b

x = float(input("Enter first number: "))
y = float(input("Enter second number: "))

print("Add =", add(x, y))
print("Subtract =", subtract(x, y))
print("Multiply =", multiply(x, y))
print("Divide =", divide(x, y))
```
